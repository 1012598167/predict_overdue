```python
mean_lend_amount=table_datamining_history_loan.groupby(['merchant_id'])['lend_amount'].sum().mean()
print(mean_lend_amount)
def func_sum_his_lend_amount(oneline):#先将在history_loan中有数据的填充好
    merchant_id=oneline['merchant_id']
    zibiao=table_datamining_history_loan[table_datamining_history_loan['merchant_id']==merchant_id]
    if not(zibiao.empty):
        return zibiao['lend_amount'].sum()
merge_loan['sum_his_lend_amount']=merge_loan.apply(func_sum_his_lend_amount,axis=1)
    #sum_send_amount=table_datamining_history_loan.groupby(['merchant_id'])['lend_amount'].sum()
#type(log_sum_send_amount)
```

```python
#补充Age的缺失值
index = list(merge_loan[merge_loan['sum_his_lend_amount'].isnull()].index) #Age 缺失样例的 index
#mean_lend_amount
copy_data = merge_loan.copy()#复制数据
for i in index:
    filling_amount = np.mean(copy_data[(copy_data['industry_level1'] == copy_data.iloc[i]['industry_level1'])
                                    & (copy_data['province'] == copy_data.iloc[i]['province'])
                                    ]['sum_his_lend_amount'])
    if not np.isnan(filling_amount): 
        merge_loan['sum_his_lend_amount'].iloc[i] = filling_amount 
    else: 
        merge_loan['sum_his_lend_amount'].iloc[i] = mean_lend_amount
del copy_data
```

```python
def func_paid_amount(oneline):
    ### 借款前30天交易总额 sum_before_30_trans_amount
    ### 前所有记录日均交易额 mean_every_day_trans_amount
    ### 日均交易额变异系数 cv_mean_every_day_trans_amount
    ### 日均交易笔数 mean_every_day_trans_count
    ### 交易总额 sum_every_trans_amount
    ### 实收金额占交易金额比例及方差(优惠程度) on_sale_rate,on_sale_var
    ### 30天内交易次数 sum_before_30_trans_count
    ### 总交易次数 sum_every_trans_count
    ### 最近一次交易距2018.6.30的天数 nearest_trans_day
    ### 笔均交易额 mean_every_trans_trans_amount
    merchant_id=oneline['merchant_id']
    zibiao=table_lend_transaction[table_lend_transaction['merchant_id']==merchant_id]
    nearest_date_delta=(datetime.datetime(2018,6,30)-zibiao['pt'].max()).days
    beforezibiao=zibiao[zibiao['pt']<oneline['apply_date']]
    starttime=beforezibiao['pt'].min()
    delta_time=1 if starttime else (oneline['apply_date']-starttime).days
    #离目标交易之前的最近一比交易时间差 如果没有流水就当是差一天
    before_every_day_transaction=beforezibiao.groupby('pt')['paid_amount'].sum()
    every_day_transaction=zibiao.groupby('pt')['paid_amount'].sum()
    #前所有记录的每日交易
    on_sale_rate=zibiao.groupby('pt')['paid_amount'].sum()/zibiao.groupby('pt')['received_amount'].sum()
    
    delta30=datetime.timedelta(days=30)
    days30=beforezibiao[beforezibiao['pt']+delta30>oneline['apply_date']]
    days30=days30.groupby('pt')['paid_amount'].sum().sum()
    return days30,before_every_day_transaction.mean(),\
        before_every_day_transaction.std()/before_every_day_transaction.mean(),\
        every_day_transaction.count(),\
        every_day_transaction.sum(),on_sale_rate.mean(),on_sale_rate.var(),\
        before_every_day_transaction.count(),every_day_transaction.count(),\
        nearest_date_delta,zibiao['paid_amount'].mean()
        
```

sma 前简单移动平均（英语：simple moving average，SMA）是某变数之前n个数值的未作加权算术平均。  
例如，收市价的10日简单移动平均指之前10日收市价的平均数。若设收市价为$$p_{1}$$至$$p_{n}$$，则方程式为： 

$$ SMA={p_{1}+p_{2}+\cdots +p_{n} \over n} $$
当计算连续的数值，一个新的数值加入，同时一个旧数值剔出，所以无需每次都重新逐个数值加起来：

$$ SMA_{t1,n}=SMA_{t0,n}-{p_{1} \over n}+{p_{n+1} \over n}$$
